```python
audio_path = "data/Red_Lines_and_Risks_in_the_AI_Act.m4a"
```

```python
# accessing and loadinf the openai key to client
from dotenv import load_dotenv
import os
from openai import OpenAI

load_dotenv()

api_key = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key=api_key)
```

```python
#creating the transcript from audio file
with open(audio_path, "rb") as audio_file:
    transcript = client.audio.transcriptions.create(
        model="gpt-4o-mini-transcribe",
        file=audio_file
    )
```

```python
podcast_text = transcript.text
```

```python
from langchain_community.document_loaders import PyPDFLoader

pdf_path = "data/385082eng.pdf"

loader = PyPDFLoader(pdf_path)
pdf_docs = loader.load()


'''Opens the PDF file.
Reads every page.
Uses pypdf to extract the text.
Creates one LangChain Document object per page.
Stores those Document objects in a Python list.
each element of a list is one page'''
```

```python
print(f"Pages loaded: {len(pdf_docs)}")
```

```python
print(pdf_docs[0].page_content[:1000])
```

```python
pdf_docs[0].metadata
```

```python
# join the pages
pdf_text = "\n\n".join(doc.page_content for doc in pdf_docs)
```

```python
print(f"Characters: {len(pdf_text):,}")

print(pdf_text[:1000])
```

```python
from langchain_text_splitters import CharacterTextSplitter
```

```python
text_splitter = CharacterTextSplitter(
    separator="\n",
    chunk_size=500,
    chunk_overlap=0
)
```

```python
text_splitter
```

```python
# test if the chunking works
fixed_chunks = text_splitter.split_documents(pdf_docs)
```

```python
print(f"Original pages: {len(pdf_docs)}")
print(f"Chunks created: {len(fixed_chunks)}")
```

```python
print(fixed_chunks[33].page_content)
```

```python
print(len(fixed_chunks[32].page_content))
# the last chunk of the page is ususlly shorter
```

```python
for i, chunk in enumerate(fixed_chunks[:3]):
    print(f"Chunk {i}")
    print(f"Length: {len(chunk.page_content)}")
    print(chunk.page_content[:100])
    print("-" * 40)

    # lot of figures, chunk size should be smaller
```

```python
# split the podcast
podcast_chunks = text_splitter.create_documents([podcast_text])
```

```python
# split the pdf
pdf_chunks = text_splitter.create_documents([pdf_text])
```
